# 02. 개요

## 2.1 이 책의 맥락

팩터 투자에서 머신러닝이 꽃을 피울 수 있었던 것은 `데이터 가용성`, `계산 능력`, `경제학적 기반`의 발전 덕분.

1. 데이터
[대체 데이터 제공업체 리스트 사이트](https://alternativedata.org/data-providers/)에서 다양한 데이터를 구할 수 있음.

2. 하드웨어와 소프트웨어를 통한 연산 능력
기술의 발전으로 데이터 처리 속도는 더 이상 기술적 장애물이 아니며, 클라우드에서 충분히 머신러닝 모델을 돌려볼 수 있음.

3. 경제학적 프레임
Arnott et al.의 종합적인 제안에 따르면, **경제학적으로 타당한 모델에 의존**하여야 한다.
이 책에서 유일하게 가정하는 것은 미래 수익률이 기업의 특성에 따라 달라진다는 것.

## 2.2 포트폴리오 구축: 작업의 흐름

성공적인 포트폴리오 전략은 여러 단계가 필요하지만, 이 책은 **예측 부분**에 중점을 둠.

> r_t+1 = f(x_t,_n) + e_t+1,_n

f함수는 t 시점에 계산한 t+1 시점의 기대수익률.

이 때, x 변수는 특성이 시가 총액, 회계 비율, 위험 측정치, 모멘텀 대용치 같은 *문헌에 보고된 고전적인 예측 인자*를 포함해야 한다는 것.

오차항 e를 2차 계획법이 가장 좋은 방식이라 생각하는 경향이 있고, 그러기에 제곱 오차를 최소화하는 전략을 사용하지만, 다른 선택지가 현명할 수 있음.

**전체적인 관점**에서 바라보는 것이 필수적.

## 2.3 머신러닝은 요술봉이 아니다

정의상, **예측의 저주**란 그 다음에 나올 가격 변동에 대한 패턴을 추론하기 위해 오직 **과거 데이터**만을 사용해야 한다는 것.

- **인과성이 핵심.** 하지만, 인과성을 발견하기는 쉽지 않음.
- 단순한 상관관계 패턴을 찾는 데 대부분의 시간을 사용하는데 이러한 패턴은 정보성이 덜하고 견고하지도 못하다.
- **무위험 차익 거래 불가 논리**는 어떤 단순한 패턴이 지속적인 수익을 창출한다면, 이 패턴은 기계적으로 그리고 매우 빠르게 사라질 것임을 주장.
- 데이터의 품질이 좌우.
- 실수로부터 배우자.
